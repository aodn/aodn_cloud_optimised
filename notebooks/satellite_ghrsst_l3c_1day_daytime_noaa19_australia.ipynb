## Access Satellite GHRSST L3C 1Day day time NOAA19 Australia (Zarr)
This Jupyter notebook demonstrates how to access and plot satellite_ghrsst_l3c_1day_daytime_noaa19_australia data, available as a [Zarr](https://zarr.dev/) dataset stored on S3.

🔗 More information about the dataset is available [in the AODN metadata catalogue](https://catalogue-imos.aodn.org.au/geonetwork/srv/eng/catalog.search#/metadata/e6782131-821c-4237-b2a9-27c6aaa8608c).

📌 The source of truth for this notebook is maintained on [GitHub](https://github.com/aodn/aodn_cloud_optimised/tree/main/notebooks/satellite_ghrsst_l3c_1day_daytime_noaa19_australia.ipynb).


```python
dataset_name = "satellite_ghrsst_l3c_1day_daytime_noaa19_australia"
```

## Install/Update packages and Load common functions

```python
import os, requests, importlib.util

open('setup.py', 'w').write(requests.get('https://raw.githubusercontent.com/aodn/aodn_cloud_optimised/main/notebooks/setup.py').text)

spec = importlib.util.spec_from_file_location("setup", "setup.py")
setup = importlib.util.module_from_spec(spec)
spec.loader.exec_module(setup)

setup.install_requirements()
setup.load_dataquery()
```

```python
from DataQuery import GetAodn
```

# Understanding the Dataset

```python
# remote zarr dataset
aodn = GetAodn()
dname = f'{dataset_name}.zarr'
%time aodn_dataset = aodn.get_dataset(dname)
```

```python
ds = aodn_dataset.zarr_store
ds.info()
```

## Plot time coverage

```python
aodn_dataset.get_temporal_extent()
```

```python
aodn_dataset.plot_time_coverage()
```

## Timeseries at a point

```python
%%time
time_series_df = aodn_dataset.get_timeseries_data(var_name='sea_surface_temperature', 
                                                  lat=-40, 
                                                  lon=130, 
                                                  date_start='2017-02-01', 
                                                  date_end='2024-10-26', 
                                                  lon_name_override="lon", 
                                                  lat_name_override="lat").plot_timeseries()
```

## Plot a gridded variable

```python
%%time
time_series_df = aodn_dataset.plot_gridded_variable(var_name='sea_surface_temperature', 
                                                    date_start='2019-06-01',                                                
                                                    lon_name_override="lon", 
                                                    lat_name_override="lat", 
                                                    lon_slice=(120, 150), 
                                                    lat_slice=(-50, -30), 
                                                    n_days=6, 
                                                    coastline_resolution="50m", 
                                                    log_scale=False)
```
